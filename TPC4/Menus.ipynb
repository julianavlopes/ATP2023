### TPC4: Aplicação para manipulação de listas de inteiros
- Crie uma aplicação em Python que coloca no monitor o seguinte menu:
    * (1) Criar Lista 
    * (2) Ler Lista
    * (3) Soma
    * (4) Média
    * (5) Maior
    * (6) Menor
    * (7) estaOrdenada por ordem crescente
    * (8) estaOrdenada por ordem decrescente
    * (9) Procura um elemento
    * (0) Sair
- O utilizador irá escolher uma das opções introduzindo o número correspondente;
- Se a opção não for sair, a aplicação executa a operação pretendida, apresenta o resultado e a seguir apresenta de novo o menu;
- Se a opção for sair, a aplicação termina colocando uma mensagem no monitor.

* No desenvolvimento da aplicação deverá ter em atenção o seguinte:
    - A aplicação terá uma variável interna para guardar uma lista de números;
    - Na opção 1, deverá ser criada uma lista de números aleatórios entre 1 e 100 que será guardada na variável interna;
    - Na opção 2, deverá ser criada uma lista com números introduzidos pelo utilizador, que será guardada na variável interna;
    - Nestas primeiras opções, se a variável interna já tiver uma lista, esta será sobreposta/apagada pela nova lista;
    - Na opção 3, será calculada a soma dos elementos na lista no momento;
    - Na opção 4, será calculada a média dos elementos na lista no momento;
    - Na opção 5, será calculado o maior elemento da lista no momento;
    - Na opção 6, será calculado o menor elemento da lista no momento;
    - Na opção 7, a aplicação deverá indicar (Sim/Não) se a lista está ordenada por ordem crescente;
    - Na opção 8, a aplicação deverá indicar (Sim/Não) se a lista está ordenada por ordem decrescente;
    - Na opção 9, a aplicação irá procurar um elemento na lista, se o encontrar deverá devolver a sua posição, devolverá -1 se o elemento não estiver na lista;
    - Se o utilizador selecionar a opção 0, a aplicação deverá terminar mostrando a lista que está nesse momento guardada.

```python
# Comando para 1:
import random
max = 100
min = 0
def criarLista(N):
    a = 0
    Lista1 = []
    while a < N:
        k = random.randint(min,max)
        Lista1.append(k)
        a = a +1
    return Lista1
# Comando para 2:
def lerLista(N):
    Lista2 = []
    a = 0
    while a < N:
        a = a + 1
        Lista2.append(int(input(">Introduza um número inteiro")))
    return Lista2
#Comando para 3:
def soma(Lista3):
    res = 0
    for x in Lista3:
        res = res + x
    return res
#Comando para 4:
def media(Lista4):
    res = 0
    for x in Lista4:
        res = res + x
    return res/len(Lista4)
#Comando para 5:
def maior(Lista5):
    res = Lista5[0]
    for x in Lista5:
        if res < x:
            res = x
    return res
#Comando para 6:
def menor(Lista6):
    res = Lista6[0]
    for x in Lista6:
        if x < res:
            res = x
    return res
#Comando para 7:
def estaOrdenadoC (Lista7):
    res = "sim"
    x = 0
    for x in range(len(Lista7) - 1):
        if Lista7[x] > Lista7[x + 1]:
            res = "nao"
    return res
#Comando para 8:
def estaOrdenadoD (Lista8):
    res = "sim"
    x = 0
    for x in range(len(Lista8) - 1):
        if Lista8[x] < Lista8[x + 1]:
            res = "nao"
    return res
#Comando para 9:
def procura (Lista9,i):
    indice = -1
    res = 0
    for res in range(len(Lista9)):
        if Lista9[res] == i:
            indice = res
    return indice
#Programa:

def menu():
    print("============MENU============")
    print("(1) Criar Lista\n(2) Ler Lista\n(3) Soma\n(4) Média\n(5) Maior\n(6) Menor\n(7) estaOrdenada por ordem crescente\n(8) estaOrdenada por ordem decrescente\n(9) Procura um elemento\n(0) Sair ")
    print("============================")
    n = int(input("Introduza um número entre 0 e 9"))
    return n


lista = criarLista(3)
ni = menu()
while ni != 0:
    if ni == 1:
        n1 = int(input("Insira um número para programa 1"))
        lista = criarLista(n1)
        print(lista)
        print("\n")
    elif ni == 2:
        n2 = int(input("Insira um número para programa 2"))
        lista = lerLista(n2)
        print(lista)
        print("\n")
    elif ni == 3:
        print(soma(lista))
        print("\n")
    elif ni == 4:
        print(media(lista))
        print("\n")    
    elif ni == 5:
        print(maior(lista))
        print("\n")
    elif ni == 6:
        print(menor(lista))
        print("\n")   
    elif ni == 7:
        print(estaOrdenadoC(lista))
        print("\n")
    elif ni == 8:
        print(estaOrdenadoD(lista))
        print("\n")
    elif ni == 9:
        n9 = int(input("Insira número que procura"))
        print(procura(lista,n9))
        print("\n") 
   
    ni = menu()
print("Obrigado por usar este menu!")

```
