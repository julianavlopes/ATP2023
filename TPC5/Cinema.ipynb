### TPC5: Aplicação para Gerir um Cinema

Suponha que está a desenvolver uma aplicacão para gestão de um conjunto de salas de cinema de um centro comercial. 
Nesse centro comercial existem algumas salas de cinema (que poderão estar a exibir filmes ou não), cada sala tem uma determinada 
lotação, uma lista com a referência dos bilhetes vendidos (lugares ocupados; cada lugar é identificado por um número inteiro), e cada sala tem um filme associado.

Considera a seguinte sugestão para o modelo dos cinemas:
```
Cinema = [Sala]
Sala = [nlugares, Vendidos, filme]
nlugares = Int
filme = String 
Vendidos = [Int]
```
  
Que poderá ser usado num programa da seguinte forma:
```
sala1 = (150, [], "Twilight")
sala2 = (200, [], "Hannibal")
cinema1 = []
...
cinema1 = inserirSala(cinema1,sala1)
cinema1 = inserirSala(cinema1,sala2)
...
listar(cinema1)
...

if(disponivel(cinema1, "Twilight", 17 )):
  cinema1 = vendebilhete(cinema1, "Twilight", 17 )
...
listardisponibilidades(cinema1)
...
```

Especifique as funções utilizadas no exemplo:

1. `listar( cinema )` - que lista no monitor todos os filmes que estão em exibição nas salas do cinema passado como argumento;
2. `disponivel( cinema, filme, lugar )` - que dá como resultado **False** se o lugar lugar já estiver ocupado na sala onde o filme está a ser exibido e dará como resultado **True** se o inverso acontecer;
3. `vendebilhete( cinema, filme, lugar )` - que dá como resultado um novo cinema resultante de acrescentar o lugar à lista dos lugares ocupados, na sala onde está a ser exibido o filme;
4. `listardisponibilidades( cinema )` - que, para um dado cinema, lista no monitor para cada sala, o filme que está a ser exibido e o total de lugares disponíveis nessa sala (número de lugares na sala menos o número de lugares ocupados);
5. `inserirSala( cinema, sala )` - que acrescenta uma sala nova a um cinema (devendo verificar se a sala já existe);
6. Acrescente todas as outras funcionalidades que achar necessárias;
7. À semelhança do exercício 3, construa uma aplicação com um menu de interface para as operações.

````python
#Verificar se existe sala
def existeSala(cinema, sala):
    nome1 = sala[2]
    i = False
    for s in cinema:
          if s[2] == nome1:
               i = True
    return i

#Inserir uma sala no cinema
def inserirSala(cinema, sala):
    if not existeSala(cinema, sala):
        cinema.append(sala)
    return cinema 

#Lista a sala com o respetivo filme
def listar(cinema):
    i = 1
    for sala in cinema:
        print(i)
        print(sala[2])
        i = i + 1

#Procura uma sala no cinema
def procuraSala(cinema, filme):
    a = [0, [], ""]
    for sala in cinema:
        if sala[2] == filme:
            a = sala
    return a

#Verifica se o lugar está disponível(True) ou ocupado(False)
def disponivel (cinema, filme, lugar):
    a = procuraSala(cinema, filme)
    if a[0] == 0:
        return False
    b = True
    for l in a[1]:
        if l == lugar:
            b = False
    return b

#Procura a posição da sala na lista do cinema
def procuraPosicao(cinema, filme):
    a = 0
    x = 0
    for sala in cinema:
        if sala[2] != filme:
            a = a + 1
        else:  
            x = a
    return x

#Se o lugar estiver disponível acrescenta-o à lista de vendidos
def vendeBilhete (cinema, filme, lugar):
    a = procuraPosicao(cinema, filme)
    if disponivel(cinema, filme, lugar) == True:
        (cinema[a][1]).append(lugar)
    else:
        print("Este lugar já está ocupado!")

#Calcula o número de lugares disponiveis
def calcularLugares (sala):
    res = 0
    nven = len(sala[1])
    res = sala[0] - nven
    return res

#Diz quantos lugares tem disponiveis na sala
def listarDisponibilidade(cinema):
    for sala in cinema:
        print(sala[2])
        print(calcularLugares(sala))

def menu():
    print("----------CINEMA----------")
    print("(1) Ver filmes em exibição")
    print("(2) Ver se o lugar está disponivel na sala")
    print("(3) Ver lugares disponíveis numa sala")
    print("(4) Acrecente o número do bilhete vendido à sala")
    print("(0)Sair")
    print("------------------------")
    n = int(input("Introduza um número de 0 a 4"))
    return n 
              
cinema1 = [[150, [1, 10, 34], "Twilight"], [200, [2, 45, 67], "Hannibal"], [170, [20,30,31,32,35,67,86,98,111], "Interstellar"], [100, [10,34,35,67,87], "Oppenheimer"], [120, [20,23,45,65,62,56], "Up: Altamente!"]]
ni = menu()
while ni != 0:
    if ni == 1:
        print(listar(cinema1))
        print("\n")
    elif ni == 2:
        f = input("Introduza o filme")
        lu = int(input("Introduza o número do lugar"))
        print(disponivel(cinema1, f, lu))
        print("\n")
    elif ni == 3:
        print(listarDisponibilidade(cinema1))
        print("\n")
    elif ni == 4:
        f = input("Introduza o filme")
        lu = int(input("Inroduza o número do bilhete vendido"))
        print(vendeBilhete(cinema1, f, lu))
        print("\n")
    ni = menu()
print("Obrigado por usar esta aplicação!")
````
